```python
%pylab inline
import numpy as np
import pickle
import glob
```

```python
loss_results = []
accuracy_results = []
file_names = glob.glob('*.pkl')
for fname in file_names:
    with open(fname, 'rb') as f:
        data = pickle.load(f)
    
    test_loss = data['test_loss']
    test_accuracy = data['test_accuracy']
    loss_results.append(test_loss)
    accuracy_results.append(test_accuracy)
```

# Plot top accuracies
