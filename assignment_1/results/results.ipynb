```python
%pylab inline
import numpy as np
import pickle
import glob
import seaborn as sns

sns.set_context("notebook", font_scale=2.5, rc={"lines.linewidth": 2.5})
sns.set_style("whitegrid")
```

```python
loss_results = []
accuracy_results = []
train_loss_results = []
train_accuracy_results = []

file_names = glob.glob('*.pkl')
for fname in file_names:
    with open(fname, 'rb') as f:
        data = pickle.load(f)
    
    # mixed up naming....
    test_loss = data['test_loss']
    test_accuracy = data['test_accuracy']
    
    
    loss_results.append(test_loss)
    accuracy_results.append(test_accuracy)
    train_loss_results.append(data['train_loss'])
    train_accuracy_results.append(data['train_accuracy'])
```
