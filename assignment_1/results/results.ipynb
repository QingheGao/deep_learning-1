```python
%pylab inline
import numpy as np
import pickle
import glob
```

```python
loss_results = []
accuracy_results = []
file_names = glob.glob('*.pkl')
for fname in file_names:
    with open(fname, 'rb') as f:
        data = pickle.load(f)
    
    train_loss = data['train_loss']
    train_accuracy = data['train_accuracy']
    loss_results.append(train_loss)
    accuracy_results.append(train_accuracy)
```

```python
plt.figure(figsize=(10,10))
for i,x in enumerate(accuracy_results):
    if x[-1] < 0.11:
        plt.plot(x, alpha=0.7,label=file_names[i])
        
plt.legend()
plt.show()
```
