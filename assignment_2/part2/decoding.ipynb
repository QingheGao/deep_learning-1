```python
import pickle
from dataset import *
```

```python
dataset = TextDataset('./books/carl_sagan.txt')

with open('./carl_sagan.txt_(rmsprop_0.002)_sampling_decoded_seqs.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
def decode(sequence, dataset):
    return "".join([dataset._ix_to_char[x] for x in sequence])
```

```python
dataset._data[:100]
```

```python
for i in range(0,20000, 200):
    seqs = data[i]
    
    batch_size = seqs.shape[1]
    
    random_batch_idx = np.random.choice(batch_size)
    print('After {} steps: '.format(i), decode(seqs[:,random_batch_idx], dataset))
```
